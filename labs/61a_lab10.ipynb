```python
(define (over-or-under num1 num2)
    (if (< num1 num2)
        -1
        (if (> num1 num2)
            1
            0
        )
    )
)

(over-or-under 1 1)
```

```python
(define (filter-lst fn lst)
    (cond ((null? lst) '())
        ((fn (car lst)) (cons (car lst) 
                              (filter-lst fn (cdr lst))))
        (else (filter fn (cdr lst)))
    )
)

(define (even? x)
  (= (modulo x 2) 0))
(filter-lst even? '(0 1 1 2 3 5 8))
```

```python
(define (make-adder num)
  (lambda (x) (+ num x))  
)

(define adder (make-adder 5))
(adder 8)
```
