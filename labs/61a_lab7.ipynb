```
def insert_into_all(item, nested_list):
  return [[item] + x for x in nested_list]

nl = [[], [1, 2], [3]]
insert_into_all(0, nl)
```

```
def subseqs(s):
  if s == []:
    return [[]]
  else:
    subset = subseqs(s[1:])
    return subset + insert_into_all(s[0], subset)

seqs = subseqs([1, 2, 3])
sorted(seqs)
```

```
def inc_subseqs(s):
  def subseq_helper(s, prev):
    if not s:
      return [[]]
    elif s[0] < prev:
      return subseq_helper(s[1:], prev)
    else:
      a = subseq_helper(s[1:], s[0]) #include  
      b = subseq_helper(s[1:], prev) #exclude
      return insert_into_all(s[0], a) + b
  return subseq_helper(s, 0)

seqs = inc_subseqs([1, 3, 2])
sorted(seqs)
```

```
def trade(first, second):
    m, n = 1, 1

    equal_prefix = lambda: sum(first[:m]) == sum(second[:n])

    def helper():
      nonlocal m, n
      while m <= len(first) and n <= len(second):
          if sum(first[:m]) == sum(second[:n]):
            return
          elif sum(first[:m]) < sum(second[:n]):
              m += 1
          else:
              n += 1
    helper()
    
    if equal_prefix():
        first[:m], second[:n] = second[:n], first[:m]
        return 'Deal!'
    else:
        return 'No deal!'

a = [1, 1, 3, 2, 1, 1, 4]
b = [4, 3, 2, 7]
c = [3, 3, 2, 4, 1]
trade(a, c) # Trades 1+1+3+2=7 for 4+3=7
```

```
def reverse(lst):
  if len(lst) > 0:
    item = lst.pop()
    reverse(lst)
    lst.insert(0, item)


original_list = [5, -1, 29, 0]
reverse(original_list)
original_list
```

```
cs61a = {
    "Homework": 2,
    "Lab": 1,
    "Exam": 50,
    "Final": 80,
    "PJ1": 20,
    "PJ2": 15,
    "PJ3": 25,
    "PJ4": 30,
    "Extra credit": 0
}

def make_glookup(class_assignments):
  curr_pts = 0
  tot_pts = 0
  def helper(ass_key, pts):
    nonlocal curr_pts, tot_pts
    curr_pts += pts
    tot_pts += class_assignments[ass_key]
    return curr_pts / tot_pts
  return helper
  
  
student1 = make_glookup(cs61a)
student1("Homework", 1.5)
student1("Lab", 1)
student1("PJ1", 18)
```

```
def num_trees(n):
  if n // 2 == 0:
    return 1
  return num_trees(n-1)

num_trees(3)
```

```
def make_advanced_counter_maker():
  global_count = 0
  def counter():
    count = 0
    def name_counter(action):
      nonlocal count, global_count
      if action == 'count':
        count += 1
        return count
      elif action == 'global-count':
        global_count += 1
        return global_count
      elif action == 'reset':
        count = 0
      elif action == 'global-reset':
        global_count = 0
    return name_counter
  return counter

make_counter = make_advanced_counter_maker()
tom_counter = make_counter()
tom_counter('count')
tom_counter('count')
tom_counter('global-count')
jon_counter = make_counter()
jon_counter('global-count')
```
