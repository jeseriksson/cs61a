```python
def eval(self, env):
  try:
    return env[self.var_name]
  except KeyError:
    raise NameError(f'name "{self.var_name}" is not defined')
```

```python
def eval(self, env):
  return self.operator.eval(env).apply([operand.eval(env) for operand in self.operands])
```
