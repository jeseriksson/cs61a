```
def make_adder_inc(a):
  def adder(b):
    nonlocal a
    add = a + b
    a = a + 1
    return add
  return adder


adder1 = make_adder_inc(5)
adder2 = make_adder_inc(6)
adder1(2)
```

```
def make_fib():
  n1 = 0
  n2 = 1
  index = -1
  def fib():
    nonlocal n1, n2, index
    index += 1
    if index == 0:
      return n1
    if index == 1:
      return n2
    res = n1 + n2
    n1 = n2
    n2 = res
    return res
  return fib

fib = make_fib()
fib()
fib()
fib()
fib()
```

```
def insert_items(lst, entry, elem):
  i = 0
  while i < len(lst):
    if lst[i] == entry:
      lst.insert(i+1, elem)
      i += 2
    else:
      i += 1
  return lst

test_lst = [1, 5, 8, 5, 2, 3]
new_lst = insert_items(test_lst, 5, 7)
new_lst
```
