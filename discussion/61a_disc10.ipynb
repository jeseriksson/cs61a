```python
(define (factorial x)
  (if (< x 2)
  1
  (* x (factorial (- x 1))))
```

```python
(define (fib n)
  (if (< n 1)
  n
  (+ (fib (- n 1)) (fib (- n 2)))
  )
)
```

```python
(define (my-append a b)
(if (null? a)
  b
  (cons (car a) (my-append (cdr a) b)))
)
```

```python
(define (insert element lst index)
(if (= index 0)
(cons element lst)
(cons (car lst) (insert element (cdr lst) (- index 1))))
)
```

```python
(define (duplicate lst)
(if (null? lst)
lst
(cons (car lst) (cons (car lst) (duplicate (cdr lst)))))
)
```
