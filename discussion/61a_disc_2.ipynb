```
def keep_ints(cond, n):
  i = 1
  while i <= n:
    if cond(i):
      print(i)
    i += 1

def is_even(x):
  return x % 2 == 0

keep_ints(is_even, 5)
```

```
def make_keeper(n):
  def do_keep(cond):
    i = 1
    while i <= n:
      if cond(i):
        print(i)
      i += 1
  return do_keep
        
def is_even(x):
  return x % 2 == 0

make_keeper(5)(is_even)
```

```
def print_delayed(x):
  def delay_print(y):
    print(x)
    return print_delayed(y)
  return delay_print

f = print_delayed(1)
f = f(2)
f = f(3)
```

```
def print_n(n):
  def inner_print(x):
    if n <= 0:
      print('done')
    else:
      print(x)
    return print_n(n-1)
  return inner_print

g = print_n(1)
g("first")("second")("third")
```
