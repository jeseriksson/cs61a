```
def multiply(m, n):
  if n == 1:
    return m
  else:
    return m + multiply(m, n-1)

multiply(5, 3)
```

```
def is_prime(n):
  if n == 1:
    return False

  def prime_helper(k):
    if k == n:
      return True

    elif n % k == 0 or n == 1:
      return False

    else:
      return prime_helper(k+1)

  return prime_helper(2)

is_prime(1)
```

```
#2 ways: 1, or 2
#base case: 1 step
#n-1 = 1 step, n-2 = 2 steps

def count_stair_ways(n):
  if n  == 1:
    return 1
  elif n == 2:
    return 2
  else:
    return count_stair_ways(n-1) + count_stair_ways(n-2)

count_stair_ways(4)
```

```
def count_k(n, k):
  if n == 0:
    return 1
  elif n < 0:
    return 0
  else:
    total, i = 0, 1
    while i <= k:
      total += count_k(n-i, k)
      i += 1
    return total

count_k(3, 3)
```

```
def even_weighted(s):
  return [i * s[i] for i in range(len(s)) if i % 2 == 0]

x = [1, 2, 3, 4, 5, 6]
even_weighted(x)
```

```
def max_product(s):
  max_1 = 0
  max_2 = 0
  for x in s:
    if x > max_1:
      max_1 = x
    else:
      if x > max_2:
        max_2 = x
  return max_1 * max_2
    

max_product([10,3,1,9,2])
```

```
def max_product_2(s):
  if s == []:
    return 1
  else:
    return max(max_product_2(s[1:]), s[0] * max_product_2(s[2:]))

max_product_2([10,3,1,9,2])
```

```
def check_hole_number(n):
  if n // 10 == 0:
    return True
  return ((n//10) % 10) < (n%10) and ((n//10)%10) < ((n//100)%10) \
    and check_hole_number(n//100)

check_hole_number(3241968)
```

```
def check_mountain_number(n):
  def helper(x, is_increasing):
    if x // 10 == 0:
      return True
    if is_increasing and (x%10) < ((x//10)%10):
      return helper(x//10, is_increasing)
    return (x%10) > ((x//10)%10) and helper(x//10, False)
  return helper(n, True)

check_mountain_number(123456)
```
