```
def square(x): 
  print("here!")
  return x * x

def so_slow(num): 
  x = num
  while x > 0:
    x=x+1
  return x / 0 
  
square(so_slow(5))
```

```
def is_prime(n): 
  """
  >>> is_prime(10)
  False
  >>> is_prime(7)
  True
  """
  if n == 1:
    return False
  else: 
    k = 2
    while k <= sqrt(n):
      if n % k == 0:
        return False
      k += 1
    return True

is_prime(7)
```
