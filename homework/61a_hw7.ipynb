```python
(define (cddr s)
  (cdr (cdr s)))

(define (cadr s)
  (car (cdr s))
)

(define (caddr s)
  (car (cdr (cdr s)))
)
```

```python
(define (sign num)
  (cond ((= 0 num) 0)
        ((< 0 num) 1)
        ((> 0 num) -1))
)
```

```python
(define (square x) (* x x))

(define (pow x y)
  (cond ((= 1 y) x)
        ((= 0 y) 1)
        ((even? y) (pow (square x) (/ y 2)))
        ((odd? y) (pow (* x (pow (square x) (/ (+ y 1) 2)))))
)
```

```python
(define (unique s)
  (if (null? s)
  nil
   (cons (car s)
   (unique (filter (lambda (x) (not (eq? x (car s)))) (cdr s)))))
)

(unique '(1 2 1 3 2 3 1))
```

```python
(define (tail-replicate x n)
  (define (iter k so-far)
    (if (= k n)
        so-far
        (iter (+ k 1) (cons x so-far))))
  (iter 0 nil)
)

(tail-replicate 2 4)
```

```python
(define (accumulate combiner start n term)
  (if (= n 0)
    start
    (accumulate combiner (combiner start (term n)) (- n 1) term))
)

(define (square x) (* x x))

(accumulate + 0 5 square)
```

```python
(define (accumulate-tail combiner start n term)
  (define (iter k so-far)
  (if (= k (+ n 1))
    so-far
   (iter (+ k 1) (combiner so-far (term k)))))
  (iter 0 start)
)

(define (square x) (* x x))

(accumulate-tail + 0 5 square)
```
