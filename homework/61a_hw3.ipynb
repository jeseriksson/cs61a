```
def g(n):
  if n <= 3:
    return n
  else:
    return g(n-1) + 2 * g(n-2) + 3 * g(n-3)

g(5)
```

```
def g_iter(n):
  if n <= 3:
    return n
  else:
    x, y, z = 1, 2, 3
    while n > 3:
      x, y, z = y, z, z+2*y+3*x
      n -= 1
  return z

g_iter(5)
```

```
def missing_digits(n):
  if n < 10:
    return 0
  else:
    if n%10 - (n%100//10) > 0:
      return missing_digits(n//10)+n%10-(n%100//10)-1
    else:
      return missing_digits(n//10)

missing_digits(15578)
```

```
def count_change(total):
  def helper(total, small_coin):
    # base case
    if total == 0:
      return 1
    elif total < 0:
      return 0
    elif small_coin > total:
      return 0
    else:
    # do include smallest coin
      with_coin = helper(total-small_coin, small_coin)

    # do not include smallest coin
      without_coin = helper(total, small_coin*2)

    # add both of those cases
      return with_coin + without_coin

  return helper(total, 1)

count_change(7)
```

```
def print_move(origin, destination):
  print("Move the top disk from rod", origin, "to rod", destination)

def move_stack(n, start, end):
  assert 1 <= start <= 3 and 1 <= end <= 3 and start != end, "Bad start/end"

  #base case
    #n = 1: move piece from start to end, finished
  if n == 1:
    return print_move(start, end)

  #recursive case
  else:
    intermediate = abs(start-end)

    #move all but the bottom piece (n-1) to the middle,
    move_stack(n-1, start, intermediate)

    #move bottom piece (1) to end,
    move_stack(1, start, end)

    #move all but the bottom piece (n-1) to the end
    move_stack(n-1, intermediate, end)
    

move_stack(3, 1, 3)
```
