```
def make_bank(balance):
  def bank(message, amount):
    nonlocal balance
    if message == 'deposit':
      balance += amount
      return print(balance)
    elif message == 'withdraw':
      if amount < balance:
        balance -= amount
        return print(balance)
      else:
        return print('insufficient funds')
    else:
      return print('invalid message')
  return bank

bank = make_bank(100)
bank('withdraw', 40)
bank('hello', 500)
bank('deposit', 20)
bank('withdraw', 90)
bank('deposit', 100)
```

```
def make_withdraw(balance, password_org):
    pass_lst = []
    def withdraw(amount, password_in):
        nonlocal balance, password_org
        if len(pass_lst) >= 3:
          return 'Too many incorrect attempts. Attempts: {0}'.format(pass_lst)
        elif password_in != password_org:
          pass_lst.append(password_in)
          return 'Incorrect password'
        else:
          if amount > balance:
              return 'Insufficient funds'
          balance = balance - amount
          return balance
    return withdraw



w = make_withdraw(100, 'hax0r')
w(25, 'hax0r')

error = w(90, 'hax0r')
error

error = w(25, 'hwat')
error

new_bal = w(25, 'hax0r')
new_bal

w(75, 'a')

w(10, 'hax0r')

w(20, 'n00b')

w(10, 'hax0r')
```

```
def repeated(t, k):
  assert k > 1
  iterator = iter(t)
  prev = next(iterator)
  counter = 1
  for x in iterator:
    if x == prev:
      counter += 1
      if counter == k:
        return x
    else:
      counter = 1
    prev = x
  return 'no value appears {0} times in a row'.format(k)


lst = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
repeated(lst, 2)
```

```
def merge(incr_a, incr_b):
  iter_a, iter_b = iter(incr_a), iter(incr_b)
  next_a, next_b = next(iter_a, None), next(iter_b, None)
  try:
    while True:

      if next_a < next_b:
        yield next_a
        next_a = next(iter_a, None)

      elif next_a > next_b:
        yield next_b
        next_b = next(iter_b, None)

      else:
        yield next_a
        next_a, next_b = next(iter_a, None), next(iter_b, None)
  
  except:
    pass 


m = merge([0, 2, 4, 6, 8, 10, 12, 14], [0, 3, 6, 9, 12, 15])
type(m)
list(m)

def big(n):
  k = 0
  while True: yield k; k += n

m = merge(big(2), big(3))
[next(m) for _ in range(11)]

```

```

def make_joint(withdraw, old_pass, new_pass):
  check = withdraw(0, old_pass)
  if type(check) is str:
    return check
  def joint(amount, pass_in):
    if pass_in == new_pass:
      return withdraw(amount, old_pass)
    return withdraw(amount, pass_in)
  return joint

w = make_withdraw(100, 'hax0r')
w(25, 'hax0r')
make_joint(w, 'my', 'secret')
j = make_joint(w, 'hax0r', 'secret')
w(25, 'secret')
j(25, 'secret')
```

```
import types

def naturals():
  i = 1
  while True:
    yield i
    i += 1

def remainders_generator(m):
  def multiples():
    i = 0
    while i < m:
      yield filter(lambda x: x % m == 1, naturals())
      i += 1
  return multiples()



[isinstance(gen, types.GeneratorType) for gen in remainders_generator(5)]

# remainders_four = remainders_generator(4)

# for i in range(4):
#   print("First 3 natural numbers with remainder {0} when divided by 4:".format(i))
#   gen = next(remainders_four)
#   for _ in range(3):
#     print(next(gen))
```
